```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Cleavland.csv')
df
```

```python
df.columns = ['age','sex','cp','trestbps','chol','fbs','restecg','thalach','exang','oldpeak','slope','ca','thal','num']
df.head()
```

```python
df.replace('?',np.nan,inplace=True)
df
```

```python
df.dropna(inplace=True)
df
```

```python
plt.figure(figsize=(10,8))
plt.hist(df['age'],bins=10,color='skyblue')
plt.xlabel('AGE')
plt.ylabel('Frequency')
plt.title('Age Distribution')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.stripplot(x='cp',y='trestbps',data=df,jitter=True)
plt.xlabel('CP')
plt.ylabel('TrestBPS')
plt.title('CP vs TrestBPS')
plt.show()
```

```python
plt.figure(figsize=(10,8))
df['num'].value_counts().plot(kind='bar',color='lightgreen')
plt.xlabel('Heart Disease')
plt.ylabel('Frequency')
plt.title('Heart Disease Distribution')
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.plot(df['thalach'])
plt.xlabel('Index')
plt.ylabel('Heart Rate')
plt.title('Heart Rate over Time')
plt.show()
```

```python
plt.figure(figsize=(10,8))
df['sex'].value_counts().plot(kind='pie',colors=['lightblue','pink'])
plt.title('Sex Distribution')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x='age',y='chol',data=df,hue='num')
plt.show()
```
